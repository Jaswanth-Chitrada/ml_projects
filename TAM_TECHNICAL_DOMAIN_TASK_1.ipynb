TAM TECHNICAL DOMAIN TASK 1



Use the dataset provided in the group and perform the following preprocessing steps-

Remove the PassengerID column because it does not provide a lot of insights.
Remove the Cabin column because it has too many missing values.
Feature scale the ‘Age’ column, both using standard scaling and min-max scaling and add the same to the dataframe.
Create a new feature ‘Cabin_detail’ using feature engineering and add it to the data frame.
Cabin_detail = Fare / Pclass

Use One Hot Encoding to encode the ‘Gender’ column
		For example, [Male = ‘0’ , Female = ‘1’]

Use Label Encoding to encode the ‘Embarked’ column
For example,[Cherbourg = ‘0 , Queenstown = ‘1’, Southampton= ‘2’]
C = Cherbourg, Q = Queenstown, S = Southampton


```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/tamvit1.csv")
df.head()
```

```python
df=df.iloc[:,1:]
df.head()
```

```python
del df['Cabin']
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler

scaler = StandardScaler()
df['Age'] = scaler.fit_transform(df['Age'].values.reshape(-1, 1))
```

```python
x_new=df['Fare']/df['Pclass']

df["cabin_detail"]=x_new

df.head()
```

```python
df = pd.get_dummies(df, columns=['Sex'], prefix=['Sex'])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder






# Perform label encoding
label_encoder = LabelEncoder()
df['Embarked'] = label_encoder.fit_transform(df['Embarked'])
df.head(20)
```
